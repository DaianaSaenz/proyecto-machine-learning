La Comunidad autonoma de Madrid esta formada solo por la provincia de madrid, tienes un csv con cada uno de los municipios de la provincia, deberas primero añadir una columna en cada dataframe señalando que pertenecen a la comunidad autonoma de Madrid y otra columna con la provincia de madrid. Una vez hecho esto deberás limpiar los datos eliminando aquellos municipios en los que el dato de renta sea nulo y eliminar en la columna con el nombre de municipio los numeros que aparece antes del nombre. Tambien cualquer cosa que veas rara. 

```python
import pandas as pd

# Leer el archivo CSV de renta_media y convertirlo en un DataFrame
df_renta_madrid = pd.read_csv('../data/raw/renta_media_hogar_ca_madrid.csv',encoding='latin-1', sep=';')
df_renta_madrid.info()

df_renta_madrid = df_renta_madrid.dropna(axis=1)
df_renta_madrid
# Eliminar la columna "Indicadores de renta media y mediana"
df_renta_madrid = df_renta_madrid.drop('Indicadores de renta media y mediana', axis=1)

# Cambiar el nombre de la columna "Total" a "Total por renta media hogar"
df_renta_madrid = df_renta_madrid.rename(columns={'Total': 'Total por renta media hogar'})

df_renta_madrid['Municipios'] = df_renta_madrid['Municipios'].str.replace('\d+', '')
df_renta_madrid['Provincia'] = 'Madrid'
df_renta_madrid['comunidad_autonoma'] = 'Comunidad de Madrid'
df_renta_madrid
```
