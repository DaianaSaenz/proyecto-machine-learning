```python
import pandas as pd
```

```python
# Leer el archivo CSV de renta_media y convertirlo en un DataFrame
df_renta_girona = pd.read_csv('../data/raw/renta_girona.csv',encoding='latin-1', sep=';')
df_renta_girona.info()

df_renta_girona = df_renta_girona.dropna(axis=1)

# Eliminar la columna "Indicadores de renta media y mediana"
df_renta_girona = df_renta_girona.drop('Indicadores de renta media y mediana', axis=1)

# Cambiar el nombre de la columna "Total" a "Total por renta media hogar"
df_renta_girona = df_renta_girona.rename(columns={'Total': 'Total por renta media hogar'})

df_renta_girona['Municipios'] = df_renta_girona['Municipios'].str.replace('\d+', '')
df_renta_girona['Provincia'] = 'Girona'
df_renta_girona['comunidad_autonoma'] = 'Cataluña'
df_renta_girona
```
