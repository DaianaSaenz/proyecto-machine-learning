```python
import pandas as pd
```

```python
# Leer el archivo CSV de municipios y total de poblacion madrid y convertirlo en un DataFrame
df_pob_municipios_madrid = pd.read_csv('../data/raw/pob_municipios_madrid.csv',encoding='latin-1', sep=';')
df_pob_municipios_madrid.info()

#limpieza del data frame y organizacion de datos
df_pob_municipios_madrid['Municipios'] = df_pob_municipios_madrid['Municipios'].str.replace(r'\d+', '')
df_pob_municipios_madrid['Provincia'] = 'Madrid'
df_pob_municipios_madrid['comunidad_autonoma'] = 'Comunidad de Madrid'
df_pob_municipios_madrid
```
