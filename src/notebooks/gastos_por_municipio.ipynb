```python
#importamos las librerias necesarias para trabajar:
import pandas as pd

# Leer el archivo CSV de gasto medio hogar por num de poblacion y convertirlo en un DataFrame
df_gasto_medio_hogar_num_pob = pd.read_csv('../data/raw/gasto_medio_hogar_num_pob.csv',encoding='latin-1', sep=';')

df_gasto_medio_hogar_num_pob = df_gasto_medio_hogar_num_pob.rename(columns={'Total': 'Media gasto total'})
df_gasto_medio_hogar_num_pob = df_gasto_medio_hogar_num_pob.drop('Tipo de dato', axis=1)
df_gasto_medio_hogar_num_pob = df_gasto_medio_hogar_num_pob.drop('Gastos totales, medios, distribución e índice', axis=1)
df_gasto_medio_hogar_num_pob["Media gasto total"] = df_gasto_medio_hogar_num_pob["Media gasto total"].str.replace('.', '', 1).str.replace(',', '.').astype(float)
df_gasto_medio_hogar_num_pob.info()
df_gasto_medio_hogar_num_pob
```

```python
# Leer el archivo CSV de gasto_medio_por_CA , convierto en un DataFrame, y organizo los datos
df_gasto_medio_hogar_por_ca = pd.read_csv('../data/raw/gasto_medio_hogar_por_ca.csv',encoding='latin-1', sep=';')
df_gasto_medio_hogar_por_ca.info()
df_gasto_medio_hogar_por_ca = df_gasto_medio_hogar_por_ca.rename(columns={'Comunidad autónoma de residencia': 'comunidad autonoma'})
df_gasto_medio_hogar_por_ca = df_gasto_medio_hogar_por_ca.drop('Tipo de hogar (reducida)', axis=1)
df_gasto_medio_hogar_por_ca = df_gasto_medio_hogar_por_ca.drop('Gastos totales, medios, distribución e índice', axis=1)
df_gasto_medio_hogar_por_ca['comunidad autonoma'] = df_gasto_medio_hogar_por_ca['comunidad autonoma'].str.replace('\d+', '')
df_gasto_medio_hogar_por_ca["Total"] = df_gasto_medio_hogar_por_ca["Total"].str.replace('.', '', 1).str.replace(',', '.').astype(float)

df_gasto_medio_hogar_por_ca
```

```python

# Obtengo
media_por_año = df_gasto_medio_hogar_por_ca.groupby('Periodo')['Total'].mean()
media_por_año

media_por_año = df_gasto_medio_hogar_por_ca.groupby("Periodo")["Total"].mean()
df_gasto_medio_hogar_por_ca["media"] = media_por_año[df_gasto_medio_hogar_por_ca["Periodo"]].values.round(2)
df_gasto_medio_hogar_por_ca
```

```python
"""calculo la diferencia entre los valores obtenidos de la tabla: gasto_medio_hogar_num_pob 
para calcular el procentaje de variacion"""

df_gasto_medio_hogar_por_ca['porcentaje_variacion'] = ((df_gasto_medio_hogar_por_ca['Total'] - df_gasto_medio_hogar_por_ca['media']) / df_gasto_medio_hogar_por_ca['media']) * 100
df_gasto_medio_hogar_por_ca
```

```python

# filtrar las filas de la tabla 1 donde la comunidad autónoma sea Madrid
tabla_madrid = df_gasto_medio_hogar_por_ca[df_gasto_medio_hogar_por_ca['comunidad autonoma'].str.contains('Madrid')]
# combinar las dos tablas en base a la columna 'Periodo'
gasto_medio_ca_madrid_num_pob = pd.merge(tabla_madrid, df_gasto_medio_hogar_num_pob, on='Periodo')
gasto_medio_ca_madrid_num_pob

def sumar_porcentaje(row):
    return row['Media gasto total'] + (  row['Media gasto total']* row['porcentaje_variacion'] / 100)

gasto_medio_ca_madrid_num_pob['Media gasto total'] = gasto_medio_ca_madrid_num_pob.apply(sumar_porcentaje, axis=1)
gasto_medio_ca_madrid_num_pob=gasto_medio_ca_madrid_num_pob.drop('Total', axis=1).drop('media', axis=1).drop('porcentaje_variacion', axis=1)
gasto_medio_ca_madrid_num_pob
```

```python


# filtrar las filas de la tabla 1 donde la comunidad autónoma sea Madrid
tabla_cataluna = df_gasto_medio_hogar_por_ca[df_gasto_medio_hogar_por_ca['comunidad autonoma'].str.contains('Catalu')]
# combinar las dos tablas en base a la columna 'Periodo'
gasto_medio_ca_cataluna_num_pob = pd.merge(tabla_cataluna, df_gasto_medio_hogar_num_pob, on='Periodo')
gasto_medio_ca_cataluna_num_pob

def sumar_porcentaje(row):
    return row['Media gasto total'] + (  row['Media gasto total']* row['porcentaje_variacion'] / 100)

gasto_medio_ca_cataluna_num_pob['Media gasto total'] = gasto_medio_ca_cataluna_num_pob.apply(sumar_porcentaje, axis=1)
gasto_medio_ca_cataluna_num_pob=gasto_medio_ca_cataluna_num_pob.drop('Total', axis=1).drop('media', axis=1).drop('porcentaje_variacion', axis=1)
gasto_medio_ca_cataluna_num_pob
```
